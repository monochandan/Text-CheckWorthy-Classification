## 2. dropping the emojis

```python
def drop_emoji(text):
    # normalize the text first
    normalized_text = unicodedata.normalize("NFKD", text)
    regex_pattern = re.compile(pattern = "["
                            u"\U0001F600-\U0001F64F"
                            u"\U0001F300-\U0001F5FF"
                            u"\U0001F680-\U0001F6FF"
                            u"\U0001F1E0-\U0001F1FF"
                            u"\U00012B07"
                            "]+", flags = re.UNICODE)
    new_text = regex_pattern.sub(r'', normalized_text)
    new_text = clean(new_text, no_emoji = True)
    return new_text
```

```python
tweet_data['after_emoji_dropped'] = tweet_data['after_link_dropped'].apply(drop_emoji)
```
