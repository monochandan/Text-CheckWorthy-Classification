```python
file = "68ecffd2jw1eqh0qh6u5lj20nq0hswg9.jpg"

# file_path = rumor_folder + "/" + file
file_path = "D:\\DATASET\\Text_Image_Checkworthy_Extn\\weibo\\rumor_images\\68ecffd2jw1eqh0qh6u5lj20nq0hswg9.jpg"

print(file_path)
```

```python
# i = 0
# keys = ['id', 'file_path', 'text', 'label']
# data_rumor = dict.fromkeys(keys, None)
# data_rumor
rumor_data = []
count = 0
```

```python
for file in os.listdir(rumor_folder):
    count += 1
    print(f"{count} of {total}")
    if file.lower().endswith(".jpg"):
        # print(file)
        # print(type(file)) ## <class 'str'>
        # i += 1
        file_path = os.path.join(rumor_folder, file)
        print(f"file path: {file_path}")
        if not os.path.exists(file_path):
            print("missing file path: ", file_path)
            continue

        # print(f"File exist for {file_path}")
        try:
            image = Image.open(file_path)
            gray_image = image.convert("L")

            text = pytesseract.image_to_string(gray_image)
            text = text.replace("\x0c", "").strip()

            if not text:
                text = "no text"
            # print(f"text from the {file}:{text}")
        except Exception as e:
            print(f"Error processing {file} and {file_path}: {e}")
            continue

        # data_rumor['id'] = file
        # data_rumor['file_path'] = file_path
        # # get the text from the image
        
        # if not text:
        #     data_rumor['text'] = "no text"
        # data_rumor['text'] = text 
        # data_rumor['label'] = 0 # rumor

        data = {
            'id': file, 
            'file_path': file_path, 
            'text': text, 
            'label': 1,
        }
        # print(f"data: {data}")
        rumor_data.append(data)
        # print(f"appended in rumor_data: {rumor_data}")
        print("==================================================")
```

```python
df_rumor = pd.DataFrame(rumor_data)
df_rumor.to_csv('rumor.csv')
```
